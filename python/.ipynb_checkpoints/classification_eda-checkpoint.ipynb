## creating df1
- filling with the average of each columns. The average_balance will be filled with the avg of the mean of q1,q2,q3,q4.


```python
df1 = df
```

```python
mean_q1 = df1['q1_balance'].loc[df1['q1_balance'].notnull()].median()
mean_q2 = df1['q2_balance'].loc[df1['q2_balance'].notnull()].median()
mean_q3 = df1['q3_balance'].loc[df1['q3_balance'].notnull()].median()
mean_q4 = df1['q4_balance'].loc[df1['q4_balance'].notnull()].median()
mean_balance = (mean_q1+mean_q2+mean_q3+mean_q4)/4
```

```python
df1['average_balance']=df1['average_balance'].fillna(mean_balance)
df1['q1_balance']=df1['q1_balance'].fillna(mean_q1)
df1['q2_balance']=df1['q2_balance'].fillna(mean_q2)
df1['q3_balance']=df1['q3_balance'].fillna(mean_q3)
df1['q4_balance']=df1['q4_balance'].fillna(mean_q4)
```
