# DEF - Functions
- run_models: run 4 different classification models;
- models_sample: apply a sampling and then run 4 different classification models;

## run_models(dataframe, y)
- This function consist in:
    - split the dataframe and y in train_test_split;
    - run a for loop applying four different methods of classification (LogisticRegression, RandomForest, KNeighbors and GradientBossting);
    - create and plot a confusion matrix (heatmap with percentages) for each method;

```python
def run_models(dataframe,y):
    classifiers = ['LogisticRegression', 'RandomForestClassifier', 'KNeighborsClassifier','GradientBoostingClassifier']
    models = [
          [LogisticRegression(max_iter = 3000), (0,0), 'LogisticRegression'],
          [RandomForestClassifier(n_estimators=200, random_state=0),(0,1), 'RandomForestClassifier'],
          [KNeighborsClassifier(), (1,0), 'KNeighborsClassifier'],
          [GradientBoostingClassifier(random_state=0),(1,1), 'GradientBoostingClassifier']
         ]
    X_train,X_test,y_train,y_test=train_test_split(dataframe,y,test_size=0.45,random_state=40)
    fig, axs = plt.subplots(2,2, figsize=(8, 8), dpi=150)       
    for i, place, name in models:
        axs[place].set_title(name)
        model = i
        model.fit(X_train,y_train)
        y_pred=model.predict(X_test)
        print(name,'accuracy is:',accuracy_score(y_test,y_pred))
        cmx=confusion_matrix(y_test,y_pred)     
        sns.heatmap(cmx/np.sum(cmx), annot=True, fmt='.3%',cmap='Blues', ax=axs[place])
        plt.title(name, fontsize = 10)
    plt.show()
```

## models_resample(dataframe, y, sample):
- This function does the same as the one above, but it does a resampling to the dataframe:
    - before run this function it's necessary to create a variable with the resampling method;

```python
def models_resample(dataframe, y, sample):    
    classifiers = ['LogisticRegression', 'RandomForestClassifier', 'KNeighborsClassifier','GradientBoostingClassifier']
    models = [
          [LogisticRegression(max_iter = 3000), (0,0), 'LogisticRegression'],
          [RandomForestClassifier(n_estimators=200, random_state=0),(0,1), 'RandomForestClassifier'],
          [KNeighborsClassifier(), (1,0), 'KNeighborsClassifier'],
          [GradientBoostingClassifier(random_state=0),(1,1), 'GradientBoostingClassifier']
         ]
    fig, axs = plt.subplots(2,2, figsize=(8, 8), dpi=150)    
    X_train,X_test,y_train,y_test=train_test_split(dataframe,y,test_size=0.45,random_state=40)
    X_sample, y_sample = sample.fit_resample(dataframe, y)
    X_train_sample,X_test_sample,y_train_sample,y_test_sample=train_test_split(X_sample,y_sample,test_size=0.45,random_state=40)
    for i, place, name in models:
        axs[place].set_title(name)
        model = i
        model.fit(X_train_sample, y_train_sample)
        y_pred_sample=model.predict(X_test)
        print(name,'accuracy is:',accuracy_score(y_test,y_pred_sample))
        cmx=confusion_matrix(y_test,y_pred_sample)     
        sns.heatmap(cmx/np.sum(cmx), annot=True, fmt='.3%',cmap='Blues', ax=axs[place])
        plt.title(name, fontsize = 10)
    plt.show()
    print('The sampling method used was:', sample,'\n')
    print('Offer Accepted on y_original: 0 = No, 1 = Yes:')
    print(y.value_counts(),'\n') 
    print('Offer Accepted on y_sample:')
    print(y_sample.value_counts(),'\n') 
    print('Offer Accepted on y_train_sample:')
    print(y_train_sample.value_counts(),'\n')
```

## models_over_under(dataframe, y, over, under):
- This function does the same as the first one with a combination of over and undersampling together:
    - before run this function it's necessary to create a variable with the over and undersampling methods;
    - after the train_test split of the dataframe, the function will do a fit_resample with the undersample, than another fit_resample on the X_under and y_under using the oversample method;
    - after applying the classification models, the funcion will show the values counts of:
        - y_original;
        - y_under;
        - y_train_und_over.

```python
def models_over_under(dataframe, y, over, under):    
    classifiers = ['LogisticRegression', 'RandomForestClassifier', 'KNeighborsClassifier','GradientBoostingClassifier']
    models = [
          [LogisticRegression(max_iter = 3000), (0,0), 'LogisticRegression'],
          [RandomForestClassifier(n_estimators=200, random_state=0),(0,1), 'RandomForestClassifier'],
          [KNeighborsClassifier(), (1,0), 'KNeighborsClassifier'],
          [GradientBoostingClassifier(random_state=0),(1,1), 'GradientBoostingClassifier']
         ]
    
    oversample = over
    undersample = under
    
    fig, axs = plt.subplots(2,2, figsize=(8, 8), dpi=150) 
    
    X_train,X_test,y_train,y_test=train_test_split(dataframe,y,test_size=0.45,random_state=40)
    X_under, y_under = undersample.fit_resample(X_train,y_train)
    X_train_und_over,y_train_und_over = oversample.fit_resample(X_under,y_under)
    
    for i, place, name in models:
        axs[place].set_title(name)
        model = i
        model.fit(X_train_und_over, y_train_und_over)
        y_pred_und_over=model.predict(X_test)
        print(name,'accuracy is:',accuracy_score(y_test,y_pred_und_over))
        cmx=confusion_matrix(y_test,y_pred_und_over)     
        sns.heatmap(cmx/np.sum(cmx), annot=True, fmt='.3%',cmap='Blues', ax=axs[place])
        plt.title(name, fontsize = 10)
    plt.show()
    print('Offer Accepted on y_original: 0 = No, 1 = Yes:')
    print(y.value_counts(),'\n') 
    print('Offer Accepted on y_under:')
    print(y_under.value_counts())
    print('Offer Accepted on y_train_und_over:')
    print(y_train_und_over.value_counts(),'\n')
```
